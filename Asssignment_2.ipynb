# 1. Write a function analyze_sentence(sentence) that takes a sentence and 
# returns a dictionary with:

```python
# word_count
# char_count (excluding spaces)
# unique_words (case-insensitive)
# most_frequent_word

# Test it with:
# "The quick brown fox jumps over the lazy dog the fox"
```

```python
def analyze_sentence(sentence):
    words=sentence.lower().split()

    word_count=len(words)
    char_count=len(sentence.replace(" ",""))
    unique_words=len(set(words)) # A set stores only unique values.

    freq={}
    for i in words:
        freq[i]=freq.get(i,0)+1
    most_freq=max(freq,key=freq.get)

    return word_count,char_count,unique_words,most_freq
```

```python
analyze_sentence("The quick brown fox jumps over the lazy dog the fox")
```

# 3. Write a function format_report(data) that takes a list of tuples 
# (name, score, grade) and prints a formatted table where:

```python

    # name is left-aligned (width 15)
    # score is right-aligned (width 6)
    # grade is centered (width 8)

# Use this data:

   #  data = [
   #    ("Alice", 92, "A"),
   #    ("Bob", 78, "B+"),
   #     ("Charlie", 55, "C"),
   #     ("Diana", 100, "A+")
   #     ]
```

```python
def format_report(data):
    print(f"{'Name':<15}{'Score':>6}{'Grade':^8}")

    for name, score, grade in data:
        print(f"{name:<15}{score:>6}{grade:^8}")

data = [
    ("Alice", 92, "A"),
    ("Bob", 78, "B+"),
    ("Charlie", 55, "C"),
    ("Diana", 100, "A+")
]

format_report(data)
```

# #4. Write a function list_ops(nums) that:


```python
# a) Removes duplicates while preserving the original order.
# b) Returns a new list with only elements at even indices, squared.
# c) Flattens the nested list into a single list using list comprehension.

#  nums = [3, 1, 4, 1, 5, 9, 2, 6, 5, 3, 5]
# nested = [[1, 2, 3], [4, 5], [6, 7, 8, 9]]
```

```python
def list_ops(nums, nested):
    unique = []
    for num in nums:
        if num not in unique:
            unique.append(num)

    even_squared = [nums[i] ** 2 for i in range(0, len(nums), 2)]
    flattened = [item for sublist in nested for item in sublist]

    return unique, even_squared, flattened


nums = [3, 1, 4, 1, 5, 9, 2, 6, 5, 3, 5]
nested = [[1, 2, 3], [4, 5], [6, 7, 8, 9]]

unique_nums, even_squared, flattened = list_ops(nums, nested)

print("unique numbers:",unique_nums)
print(" squared number:",even_squared)
print("flattened number",flattened)
```

```python

# a) Unpack it into meaningful variable names.

# b) Use * unpacking to separate the first element, the last, and everything in between.

# c) Write a function update_salary(emp_tuple, new_salary) that returns a new tuple with the salary updated.
```

```python
employee = ('E042', 'Riya Sharma', 'Data Science', 72000, 'Pune')

# a) Unpack into meaningful variable names
emp_id, name, department, salary, city = employee

print("Employee ID:", emp_id)
print("Name:", name)
print("Department:", department)
print("Salary:", salary)
print("City:", city)


# b) Use * unpacking 
first, *middle, last = employee

print("\nFirst Element:", first)
print("Middle Elements:", middle)
print("Last Element:", last)


# c) Function to update salary
def update_salary(emp_tuple, new_salary):
    emp_id, name, department, salary, city = emp_tuple
    return (emp_id, name, department, new_salary, city)

# Function call
updated_employee = update_salary(employee, 80000)

print("\nOriginal Tuple:", employee)
print("Updated Tuple:", updated_employee)
```

# 7. Write a function inventory_manager(records) that takes a list of tuples (product, category, quantity, price) and  returns a dictionary grouped by category.


```python

# Each category should have:

# total_items
# total_value
# A list of product names

# Use:

# records = [
#    ('Laptop', 'Electronics', 10, 55000),
#    ('Mouse', 'Electronics', 50, 499),
#    ('Desk', 'Furniture', 8, 12000),
#    ('Chair', 'Furniture', 15, 4500),
#    ('Pen', 'Stationery', 200, 20)
# ]
```

```python
records = [
    ('Laptop', 'Electronics', 10, 55000),
    ('Mouse',   'Electronics', 50, 499),
    ('Desk', 'Furniture', 8, 12000),
    ('Chair', 'Furniture', 15, 4500),
    ('Pen', 'Stationery', 200, 20)
]

def inventory_manager(records):
    inventory = {}

    for product, category, quantity, price in records:
        if category not in inventory:
            inventory[category] = {
                'total_items': 0,
                'total_value': 0,
                'products': []
            }

        inventory[category]['total_items'] += quantity
        inventory[category]['total_value'] += quantity * price
        inventory[category]['products'].append(product)

    return inventory


# Function call
result = inventory_manager(records)

# Display result
for category, details in result.items():
    print(f"\nCategory: {category}")
    print("Total Items:", details['total_items'])
    print("Total Value:", details['total_value'])
    print("Products:", details['products'])
```
