```python
year = int(input("Enter a year: "))

if year % 4 == 0:
    if year % 100 == 0:
        if year % 400 == 0:
            print("Leap year")
        else:
            print("century but Not a leap year")
    else:
        print("Leap year but not century")
else:
    print("Not a leap year")
                    
```

```python
"""
menu driven ->
take a string from the user
menu-> upper lowercase , replace a particular part of the string 
for third option it should be which part to replace with that
"""
```

```python
str1=input("enter the string:")
print("menu=> \n.upper case \n2.lowercase \n3.replace ")

ch=int(input("enter the choice:"))
if ch==1:
    print("result:",str1.upper())
elif ch==2:
    print("result:",str1.lower())
elif ch==3:
    old=input("enter the part to replace:")
    new=input("enter the text:")
    if old in str1:
        print("Result:", str1.replace(old, new))
    else:
        print("Part not found in string")

else:
    print("Invalid choice")
    
    
```
