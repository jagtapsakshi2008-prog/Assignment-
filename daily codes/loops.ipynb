```python
# find a factorial of a number using while loop
```

```python
fact=1
i=1
num=int(input("enter number:"))
while i<=num:
    fact=fact*i
    i=i+1
print("factorial:",fact)
```

```python
# find the addition of digits of a number
# eg-> 12345 => 1+2+3+4+5=15
#hint -> use floor division and modulus
```

```python
num=12345
sum=0
while(num>0):
    sum=sum+num%10
    num=num//10
print("sum of digits:",sum)
```

```python
num=5
fact=1
while num>0:
    fact=fact*num
    num=num-1
print("factorial:",fact)
```
