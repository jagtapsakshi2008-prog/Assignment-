```python
class person:
    def __init__(self,name,age,ht):
        self.name=name
        self.__age=age
        self.ht=ht
    def show_info(self):
        print("name=>",self.name)
        print("age=>",self.__age)
        print("height=>",self.ht)
```

```python
"""
create a class called employee
attributes => company,name,monthly_salary,job_designation
here all the objects of this company will have same common company
make salary as private variable

create 3 objects -> e1,e2,e3 with given attributes 

method 1=> method to calculate annual salary
method 2=> method to show entire data

"""
```

```python
class Employee:
    company="ABC Pvt Ltd" 
    def __init__(self,name,mon_sal,job_designation):
        self.company=Employee.company
        self.name=name
        self.__mon_sal=mon_sal
        self.job_designation=job_designation

    def cal_annual_salary(self):
        annual_sal=self.__mon_sal*12
        return annual_sal

    def show(self):
        print("company=>",self.company)
        print("name=>",self.name)
        print("monthly salary=>",self.__mon_sal)
        print("job designation=>",self.job_designation)
        print("annual salary=>",self.cal_annual_salary())
        
    
    
```

```python
e1=Employee("john",45000,"data analyst")
e2=Employee("smith",30000,"project manager")
e3=Employee("tony",40000,"software engineer")
e1.show()
```
