```python
class Student:
    def __init__(self,name,age):
        self.name=name
        self.age=age
        self.skills=[]

    def Student_skill(self,skill):
        self.skills.append(skill)

    def show_info(self):
        print("name=>",self.name)
        print("age=>",self.age)
        print("skills learned by a student=>",self.skills)
        
        
        
```

# Inheritance

```python
class Person:
    insti="MITU"
    def __init__(self,age,ht):
        self.age=age
        self.ht=ht

    def show(self):
        print(f"age=>{self.age},height=>{self.ht},instute=>{self.insti}")
```

```python
class Soham(Person):
    def __init__(self,age,ht,sal):
        super().__init__(age,ht)
        self.sal=sal

    def show_sal(self):
        print(f"salary is =>{self.sal}")
```

```python
s=Soham(24,4.5,90000)
s.show_sal()
s.show()
```

```python
class Menuapp:
     
     
    def __init__(self):
        self.itemlist=[]

    def show_menu(self):
        print("Menu\n1.add items\n2.view items\n3.exit")

    def add_items(self):
        item=input("enter the item name :")
        self.itemlist.append(item)

    def view_items(self):
        return self.itemlist

    def run(self):
        while True:

            self.show_menu()
            ch=int(input("enter your choice:"))

            if ch==1:
                self.add_items()
            elif ch==2:
                self.view_items()
            elif ch==3:
                print("exit")
                break
            else:
                print("invalid choice")
soham = Menuapp()
soham.run()
```

```python
class Person:
    def __init__(self,age,name):
        self.__age=age
        self.__name=name

    # getter
    def get_name(self):
        return self.__name

    def get_age(self):
        return self.__age

    # getter
    def set_age(self,age):
        self.__age=age

    def __str__(self):
        return f"age=>{self.__age},name=>{self.__name}"
```

```python
# create a menu driven app for banking simualtion
# check balance, deposit money , withdraw money 
# ( you can add more things as you see fit to make the app better
```

```python
class Bankapp:
    def __init__(self,name,balance=0):
        self.name=name
        self.balance=balance

    def check_balance(self):
        print("current balance=>",self.balance)

    def deposit(self):
        amount=float(input("enter the amount to deposit:"))
        self.balance=self.balance+amount
        print("amount deposited successfully>.!!")

    def withdraw(self):
        amount=float(input("Enter amount to withdraw"))

        if amount<=self.balance:
            self.balance=self.balance-amount
            print("withhdrawal successfully..!!")
        else:
            print("insufficient balance")

    def show_details(self):
        print("\n account name=<",self.name)
        print("balance=>",self.balance)

    def run(self):
        while(True):
            print("\n Bank Menu")
            print("1. check balance")
            print("2. deposit money")
            print("3. withdraw money")
            print("4. show account details ")
            print("5. Exit")

            ch=int(input("enter the choice"))

            if ch==1:
                self.check_balance()
            elif ch==2:
                self.deposit()
            elif ch==3:
                self.withdraw()
            elif ch==4:
                self.show_details()
            elif ch==5:
                print("Exit")
                break
            else:
                print("invalid choice")
                
                
        
    
```

```python
b=Bankapp("Soham",30000)
```

```python
b.run()
```
