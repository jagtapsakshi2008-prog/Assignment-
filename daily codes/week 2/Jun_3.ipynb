```python
""""
take a string as a user input
use aloop and if-else to check if a specific character is present in the string or not.
get output for both if the character is present and if it not present.

user input->string,character1,character2
"""
```

```python
str1=input("enter the string:")
char1=input("enter the character : 1")
char2=input("enter the character : 2")
found1=False
found2=False
for i in str1:
    if i==char1:
        found1=True
    if i==char2:
        found2=True

if found1==True:
    print(f"'{char1}' is present")
else:
    print(f"'{char1}' is not present.")

if found2==True:
    print(f"'{char2}' is present")
else:
    print(f"'{char2}' is not present.")
    
    
```
