```python
# take a year as input and pritn whether it is leap year or not .
```

```python
year=int(input("enter the year:"))

if year % 4 == 0:
    if year % 100 == 0:
        if year % 400 == 0:
            print("Leap year")
        else:
            print("century but Not a leap year")
    else:
        print("Leap year but not century")
else:
    print("Not a leap year")
```

```python
# 6. build a simple calculator - 
# take two number and an operator(+,-,*,/) as input perform the operation, and print the result .
# print an error message for invalid operators or division by zero .
```

```python
print("welcome to simple calculator:")
num1=int(input("enter the first number:"))
num2=int(input("enter the second number:"))
operator=input("enter the operator:")

if(operator=="+"):
    print("addition=",num1+num2)
elif(operator=="-"):
    print("subtraction=",num1-num2)
elif(operator=="*"):
    print("multiplication=",num1*num2)
elif(operator=="/"):
    if(num2==0):
        print("Error: division by zero is not allowed")
    else:
        print("division=",num1/num2)

else:
    print("Error: invalid  operator")
```

```python
# 8. take a integer n as input and print all prime numbers from 2 to n.
```

```python
n = int(input("Enter a number: "))

for num in range(2, n + 1):
    is_prime = True

    for i in range(2, num):
        if num % i == 0:
            is_prime = False
            break

    if is_prime:
        print(num)
```

```python
# 9. keep accepting numbers from the user until they enter 0. 
# then print the total count of number entered , their sum, and their average.
# Do not include 0 in the calculations.
```

```python
# Accept numbers until user enters 0

count = 0
total = 0

while True:
    num = int(input("Enter a number (0 to stop): "))

    if num == 0:
        break

    count += 1
    total += num

# Calculate average
if count > 0:
    average = total / count
else:
    average = 0

# Print results
print("Count =", count)
print("Sum =", total)
print("Average =", average)
```

```python
# 10. take n as input and print a triangle pattern of n rows made of asterisks,
# where row i has(2i-1) stars centered appropriately.
```

```python
n = int(input("Enter number of rows: "))

for i in range(1,n+1):
    spaces=n-i
    stars=2*i-1

    print(" " *spaces + "*" *stars)
```
