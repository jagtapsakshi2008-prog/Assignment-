```python
# 9. Create a module stats_utils.py with three functions:

# a) mean(data) — arithmetic mean.

# b) variance(data) — population variance without using the statistics module.

# c) normalize(data) — min-max normalization returning values between 0 and 1.

# In a separate main.py, import and use these functions on the dataset below. Also compute the standard deviation using the math module.

# dataset = [12, 45, 7, 23, 56, 89, 34, 67, 11, 90]
```

```python
# stats_utils.py

def mean(data):
    return sum(data) / len(data)

def variance(data):
    m = mean(data)
    return sum((x - m) ** 2 for x in data) / len(data)

def normalize(data):
    min_val = min(data)
    max_val = max(data)
    return [(x - min_val) / (max_val - min_val) for x in data]
```

```python
import math

dataset = [12, 45, 7, 23, 56, 89, 34, 67, 11, 90]

# Mean
m = mean(dataset)
print("Mean =", m)

# Variance
v = variance(dataset)
print("Variance =", v)

# Normalize
n = normalize(dataset)
print("Normalized Data =", n)

# Standard Deviation
std_dev = math.sqrt(v)
print("Standard Deviation =", std_dev)
```
