```python
# 1. Create a class BankAccount with the following:

# a) Attributes: account_holder, account_number, and balance.

# b) Methods:

# deposit(amount)
# withdraw(amount)
# get_balance()

# c) withdraw() should print a message and do nothing if the amount exceeds the current balance.

# d) A __str__ method that returns a readable summary of the account.

# Create two BankAccount objects and perform a few deposits and withdrawals on each.
```

```python
class BankAccount:
    def __init__(self,acc_holder,acc_number,balance=0):
        self.acc_holder=acc_holder
        self.acc_number=acc_number
        self.balance=balance

    def deposit(self):
        amount=int(input("enter the amount to deposit:"))
        self.balance=self.balance=amount
        print(f"{amount} is deposited successfully")

    def withdraw(self):
        amount=int(input("enter the amount to withdraw:"))
        if amount<=self.balance:
            self.balance=self.balance-amount
            print(f"{amount} is withdraw successfully")
        else:
            print("unsufficient balance")
    
    def get_balance(self):
        print("current balance=",self.balance)

    def __str__(self):
        print(f"account holder =>{acc_holder}")
        print(f"account number =>{acc_number}")
        print(f"balance =>{balance}")


```

```python
# 2. Create a class Student with attributes:
# name
# roll_no
# marks (a list of 5 subject marks)

# Add methods:

# a) calculate_percentage() — returns the average of all marks.

# b) get_grade() — returns:

# 'A' for percentage >= 80
# 'B' for percentage >= 60
# 'C' for percentage >= 40
# 'Fail' otherwise

# c) display() — prints all details including percentage and grade.
# Create a list of 3 Student objects and call display() on each.
```

```python
class Student:
    def __init__(self,name,roll_no,marks):
        self.name=name
        self.roll_no=roll_no
        self.marks=marks

    def calculate_percentage(self):
        return sum(self.marks)/len(self.marks)

    def get_grade(self):
        percentage=self.calculate_percentage()

        if percentage>=80:
            return "A"
        elif percentage>=60:
            return "B"
        elif percentage>=40:
            return "C"
        else:
            return "Fail"

    def display(self):
        print("name =>",self.name)
        print("roll no =>",self.roll_no)
        print("marks =>",self.marks)
        print("percentage =>",self.calculate_percentage())
        print("grade => ",self.get_grade())
        print("\n")
```

```python
s1=Student("soham",45,[89,90,96,95,93])
s2=Student("sakshi",8,[97,87,86,94,92])
s3=Student("Sam",4,[89,86,88,98,91])
```

```python
students=[s1,s2,s3]
for i in students:
    i.display()
```

# Section B — Encapsulation

```python
# 3. Create a class Employee with private attributes:
# __name
# __emp_id
# __salary

# Add:

# a) Getter and setter methods for all three attributes.
# b) The salary setter should raise a ValueError if the value is below 10000.
# c) A method give_raise(percent) that increases salary by the given percentage.

# Demonstrate that the private attributes cannot be accessed directly, and 
# that all access goes through the methods.
```

```python
class Employee:
    def __init__(self,name,emp_id,salary):
        self.__name=name
        self.__emp_id=emp_id
        self.set_salary(salary)
        
    def get_name(self):
        return self.__name

    def get_emp_id(self):
        return self.__emp_id

    def get_salary(self):
        return self.__salary

    def set_name(self,name):
        self.__name=name

    def set_emp_id(self,emp_id):
        self.__emp_id=emp_id

    def set_salary(self,salary):
        if salary<10000:
            self.__salary=salary

    def give_raise(self,percent):
        self.__salary+=self.__salary*percent/100

    def display(self):
        print("name =>",self.__name)
        print("employee ID =>",self.__emp_id)
        print("salary =>",self.__salary)

    
    
```

```python
e1=Employee("soham",101,2500)
```

```python
print("name =>",e1.get_name())
print("employee ID =>",e1.get_emp_id())
print("salary =>",e1.get_salary())
```

```python
# give 10% raise
e1.give_raise(10)
print("after 10% raise")
e1.display()
e1.set_salary(5000)
print("\n after salary updates")
e1.display()
```

```python
# 4. Build a class SecureVault that stores a private __pin and a private __data dictionary.
# It should have:

# a) A method add_entry(pin, key, value) that adds to __data only if the pin matches.
# b) A method get_entry(pin, key) that returns the value only if the pin matches.
# c) Any wrong pin attempt should print "Access Denied" and return nothing.
```

```python
class SecureVault:
    def __init__(self,pin):
        self.__pin=pin
        self.__data={}

    def add_entry(self,pin,key,value):
        if pin==self.__pin:
            self.__data[key]=value
            print("Entry added successfully")
        else:
            print("access denied")

    def get_entry(self,pin,key):
        if pin==self.__pin:
            return self.__data.get(key)
        else:
            print("access denied")
            return None
 
```

```python
v1=SecureVault(1234)
v1.add_entry(1234,"email","soham@gmail.com") # add entry
v1.add_entry(1234,"phone","3467128990")

# wrong pin
v1.add_entry(2212,"address","pune")

print("Email =>",v1.get_entry(1234,"email")) # get entry
print("phone =>",v1.get_entry(1234,"phone"))

# wrong pin
print(v1.get_entry(2212,"address"))
```

# Section C — Inheritance and Method Overriding

```python
# 5. Create a base class Shape with:
# A method area() that returns 0.
# A method describe() that prints "I am a shape".

# Create three subclasses:

# a) Circle(radius) — overrides area() using: π × r²

# b) Rectangle(length, width) — overrides area().

# c) Triangle(base, height) — overrides area().

# Each subclass should also override describe() to print something specific about itself.

# Create one object of each class and call both methods (area() and describe()).
```

```python
import math

class Shape:
    def area(self):
        return 0

    def describe(self):
        print("I am shape")

class Circle:
    def __init__(self,radius):
        self.radius=radius

    def area(self):
        return math.pi*self.radius**2

    def describe(self):
        print("I am a circle")

class Rectangle:
    def __init__(self,length,width):
        self.length=length
        self.width=width

    def area(self):
        return self.length*self.width

    def describe(self):
        print("\n I am rectangle")

class Triangle:
    def __init__(self,base,height):
        self.base=base
        self.height=height

    def area(self):
        return 0.5*self.base*self.height

    def describe(self):
        print("\n I am Triangle")
        
```

```python
c1=Circle(7)
r1=Rectangle(10,5)
t1=Triangle(8,6)

c1.describe()
print("area =>",c1.area())

r1.describe()
print("area =>",r1.area())

t1.describe()
print("area =>",t1.area())
```

```python
# 6. Create a class Vehicle with attributes:
# make,model,year,and a method start_engine() that prints:
# Engine started.

# Create two subclasses:

# a) Car(make, model, year, num_doors) — overrides start_engine() to include the car's name in the message.
# b) ElectricCar(make, model, year, battery_capacity) — overrides start_engine() to print:
# Silent start — battery at X kWh.

# Use super() in both subclasses to call the parent __init__.
# Create objects and test all methods.
```

```python
class Vehicle:
    
    def __init__(self,make,model,year):
        self.make=make
        self.model=model
        self.year=year

    def start_engine(self):
        print("engine started")


class Car(Vehicle):
    
    def __init__(self,make,model,year,num_doors):
        super().__init__(make,model,year)
        self.num_doors=num_doors

    def start_engine(self):
        print(f"Engine started for =>{self.make} {self.model}.")


class ElectricCar(Vehicle):
    
    def __init__(self,make,model,year,battery_capacity):
        super().__init__(make,model,year)
        self.battery_capacity=battery_capacity

    def start_engine(self):
        print(f"Silent start — battery at=>{self.battery_capacity} X kWh")
        
```

```python
v1=Vehicle("Royal enfield","city",2022)
c1=Car("XUV3xO","coralla",2024,4)
e1=ElectricCar("tesla","model 3",2025,75)
```

```python
v1.start_engine()
c1.start_engine()
e1.start_engine()
```

```python
# 8. Create a class Calculator with overloaded-style methods using default arguments:
# a) add(a, b, c=0) — adds two or three numbers.b) multiply(a, b, c=1) — multiplies two or three numbers.c) 
# Also implement __add__ on a class Vector(x, y) so that two Vector objects can be added with the + operator.
# Print the resulting vector.
```

```python
class Calculator:
    
    def add(self,a,b,c=0):
        return a+b+c

    def multiply(self,a,b,c=1):
        return a*b*c

class Vector:
    def __init__(self,x,y):
        self.x=x
        self.y=y

    def __add__(self,other):
        return Vector(self.x+other.x, self.y+other.y)

    def __str__(self):
        return f"Vector({self.x} , {self.y})"

calc=Calculator()
print("Addition of 2 numbers=>",calc.add(10,20))
print("Addition of 3 numbers=>",calc.add(10,20,30))

print("Multiplication of 2 numbers=>",calc.multiply(5,4))
print("multiplication of 3 numbers=>",calc.multiply(5,4,2))

```

```python
v1=Vector(2,3)
v2=Vector(4,5)

result=v1+v2
print("Resulting Vector:",result)
```
